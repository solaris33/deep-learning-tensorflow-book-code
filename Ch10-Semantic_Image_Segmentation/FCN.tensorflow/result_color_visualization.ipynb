```python
import scipy.misc as misc
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
%matplotlib inline
```

```python
gt1 = misc.imread('logs/gt_5.png')
inp1 = misc.imread('logs/inp_5.png')
pred1 = misc.imread('logs/pred_5.png')
gt2 = misc.imread('logs/gt_6.png')
inp2 = misc.imread('logs/inp_6.png')
pred2 = misc.imread('logs/pred_6.png')
```

```python
plt.figure()
plt.imshow(inp1)
plt.figure()
plt.imshow(inp2)
```

```python
plt.figure()
plt.imshow(gt1, cmap=cm.Paired ,vmin=0, vmax=151)
plt.colorbar()
plt.savefig('logs/gt_5_color.png')
plt.figure()
plt.imshow(gt2, cmap=cm.Paired ,vmin=0, vmax=151)
plt.colorbar()
plt.savefig('logs/gt_6_color.png')
```

```python
plt.figure()
plt.imshow(pred1,cmap=cm.Paired ,vmin=0, vmax=151)
plt.colorbar()
plt.savefig('logs/pred_5_color.png')
plt.figure()
plt.imshow(pred2,cmap=cm.Paired ,vmin=0, vmax=151)
plt.colorbar()
plt.savefig('logs/pred_6_color.png')
```
